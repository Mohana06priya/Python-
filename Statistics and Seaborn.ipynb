### Measure of Central Tendency

'mean: sum of observations / No. of observations [observations --> rows]

```python
df = pd.DataFrame({"Sample_data":[1,2,4,7,9,8,11,5]})
df
```

```python
# mean

df['Sample_data'].mean()
```

```python
# median

df['Sample_data'].median()
```

```python
df = pd.DataFrame({"Sample_data":[1,2,4,7,9,8,11,5,14]})
df['Sample_data'].median()
```

```python
# Mode:
df = pd.DataFrame({"Sample_data":[1,2,4,7,9,8,11,5,14,1]})
a = df['Sample_data'].mode()
```

## Correlation

```python
# so correlation can be done using a technique called heatmap:

```

```python
titanic_data.head()
```

```python
df = titanic_data[['Age','Fare']]
df.head()
```

```python
df.corr(method='pearson')
```

```python
# heatmap:

sns.heatmap(data=df.corr(method='pearson'), vmin=-1 , vmax=1, annot=True, cmap='tab20')
```

```python
titanic_data.head()
```

```python
data=titanic_data[['PassengerId','Survived','Pclass','Age','SibSp','Parch','Fare']]
data
```

```python
# apply heat map for entire data

# heatmap:

sns.heatmap(data = data.corr(method='pearson'), vmin = -1,vmax=1,annot=True,cmap='tab20')
```

```python
## spearman

sns.heatmap(data = data.corr(method='spearman'), vmin = -1,vmax=1,annot=True,cmap='tab20')
```
