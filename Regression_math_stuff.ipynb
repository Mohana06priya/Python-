### what is the difference between R2_score and Adjusted R2_score

```python
# with proper data
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('case2.csv')
df.head()
```

```python
x = df.iloc[: , 0]
y = df.iloc[: , -1]
x = x.values.reshape(-1,1)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y) # the model will develop

pred = reg.predict(x) # prediction[real time we call them as inference]

res = pd.DataFrame({'Actual':y , 'predictions': pred})
res
```

```python
# find the accuracy for this model: [with best data we got 70% accuracy]

from sklearn.metrics import r2_score
r2_score(y , pred)
```

```python
len(df) # number of rows
```

```python
df.shape[1]  # gives the number of columns
```

```python
# find the adjusted r2_score for this model:
# adjusted r2_score is used to find the accuracy as well as we can decide the new column is important or not
# for existing data set

# finding using formula

1- (1-r2_score(y,pred)) * (len(df)-1) / (len(df) - df.shape[1] - 1)

# adjusted r2_score value is 65%
```

```python
df = pd.read_csv('case.csv')
df.head()
```

```python
x = df.drop(['Salary'] , axis = 1)
y = df['Salary']

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y) # the model will develop

pred = reg.predict(x) # prediction[real time we call them as inference]

res = pd.DataFrame({'Actual':y , 'predictions': pred})
res
```

```python
from sklearn.metrics import r2_score
r2_score(y , pred)
```

```python
1- (1-r2_score(y,pred)) * (len(df)-1) / (len(df) - df.shape[1] - 1)
# adjusted r2_score for this model is 66%
```

```python
# split the dataset:

x = df.iloc[: , -1]
y = df.iloc[: , -1]

# we can implement on top of entire data because we are not testing anything

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state  =42)

# give the data to the model

from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(X_train,y_train)  # model will be develop here 

# now find the test accuracy of the model

y_test_pred = reg.predict(X_test)

# find the accuracy of the test data to check the performance
from sklearn.metrics import r2_score
print(f'The accuracy of test data :{r2_score (y_test , y_test_pred)}')
```
