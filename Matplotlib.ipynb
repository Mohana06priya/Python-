### Scatterplot

```python
# Scatterplot is used to visualize the data in only 2-d format
```

```python
# reduce the output image size

plt.figure(figsize=(5,3)) # (width, height) this line should be in the first line itself

x = np.array([1,24,6,11,8,2,7]) # this is 1d
y = np.array([3,2,1,6,8,11,16])  # this is 1d

# using matplotlib visualize the data

# syntax = scatter(x_axis_values,y_axis_values)
plt.scatter(x,y, marker = '*', color = 'y')
plt.title('Scatterplot')
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.xticks([])
plt.yticks([])
plt.show()   # for seeing the image 
```

```python
# dont do this mistake:

x = np.arange(1,21,2)
y = np.arange(1,21,3)

print(x)
print(y)
```

```python
# if it is a scalar value(in list)

# its is not a good approach to give list values for visualization , use 1 d data for scatter point
plt.figure(figsize=(5,3))

x = [1,24,6,11,8,2,7] 
y = [3,2,1,6,8,11,16]

plt.scatter(x,y, marker = '*', color = 'y')

plt.show()  
```

### Lineplot just join the plots:

```python
plt.figure(figsize=(5,3)) # (width, height) this line should be in the first line itself

x = np.array([1,24,6,11,8,2,7]) # this is 1d
y = np.array([3,2,1,6,8,11,16])  # this is 1d

# using matplotlib visualize the data

# syntax = scatter(x_axis_values,y_axis_values)
plt.plot(x,y, marker = '*', color = 'y')

plt.show()   # for seeing the image 
```

### Legend : it tries to clarify the doubt

```python
# used to find the data belongings:

co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
c = [45,48,48,46,40,42,41]


plt.figure(figsize=(5,3))

plt.plot(co,a , label = 'C0_A')
plt.plot(co,b , label = 'CO_B')
plt.plot(co,c , label = 'CO_C')


plt.legend(loc = 4 , fontsize = 'small')
plt.show()
```

### Histogram*****

```python
a = [10,11,14,18,19,22,24,36,38,49]
plt.figure(figsize=(5,3))
plt.hist(a,bins = 5)
plt.show()
```

### Subplot

```python
# Subplots : It means we can do the visualizations differently:
plt.figure(figsize=(9,7))


co = [1,2,3,4,5,6,7]
a = [50,51,52,48,47,49,46]
b = [43,42,40,44,33,35,37]
c = [45,48,48,46,40,42,41]


# All in one graph:
plt.subplot(2,2,1)

plt.plot(co,a , label = 'C0_A')
plt.plot(co,b , label = 'CO_B')
plt.plot(co,c , label = 'CO_C')
plt.title('3 combinations')

# second:
plt.subplot(2,2,2)
plt.plot(co,a , color = 'g')
plt.title('co a')
#third
plt.subplot(2,2,3)
plt.plot(co,b , color = 'r')
plt.title('co b')
#fourth
plt.subplot(2,2,4)
plt.plot(co,c , color = 'black')
plt.title('co c')
plt.show


```

### Bar_chart

```python
categories = ['a' , 'b' , 'c' , 'd']
values = [25,50,75,30]
plt.figure(figsize=(3,3))
plt.bar(categories,values)
plt.grid(True)
plt.xlabel('categories')
plt.ylabel('values')
plt.show
```

```python
cities = ['New York','Tokyo','London','Paris','Sydney']
population=[8623000,9273000,8908081,2140526,5312000]
colours=['blue','orange','green','red','purple']
plt.figure(figsize=(4,4))
plt.bar(cities, population,color=colours)
plt.show()
```

### Pie chart

```python
languages = ['Python','Java','C++','Javascript','C#']
popularity = [30,25,15,10,20]
a=['blue','orange','green','red','purple']
explode = [0.1,0.3,0.4,0.2,0.3]
plt.pie(popularity , labels = languages , colors=a , autopct='%1.1f%%',shadow = True, explode=explode)
plt.show()
```
